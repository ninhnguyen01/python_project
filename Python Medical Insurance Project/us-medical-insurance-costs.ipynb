Goals: <br>

The insurance data comes from the file "insurance.csv" which consists of 1338 observations of 7 variables.

Variable Descriptions: <br>

age: age in years <br>
sex: male or female <br>
bmi: body mass index <br>
children: do they have children? (number of children) <br>
smoker: do they smoke? (yes or no) <br>
region: where the person is located (eg., southwest) <br>
charges: the insurance cost for the individual <br>

## 1) The average age of the patients in the dataset.

```python
# import csv module
# reading the csv file after conversion to dictionary
import csv

with open('insurance.csv') as insurance:
    insurance_reader = csv.DictReader(insurance)
    for i in insurance_reader:
        print(i)
```

```python
# pandas and seaborn modules imported

import pandas as pd
import seaborn as sns
```

```python
# reading csv file 

insurance = pd.read_csv('insurance.csv')
insurance
```

```python
# examining number of observations and variable types 

insurance.shape
```

```python
# selecting the 'age' variable

insurance['age']
```

```python
# plotting a histogram of the age variable

age = sns.histplot(x = 'age', data = insurance)
```

```python
# describing the age variabe to find the average patient age

insurance['age'].describe()
```

The average patient age is 39.

## 2) Where a majority of the individuals are from.
  


```python
# using the max function to find the top region

insurance['region'].max()
```

The majority of people are from the southwest region.

## 3) The different costs between smokers vs. non-smokers.


```python
count = sns.countplot(x = 'smoker', data = insurance)
```

```python
insurance['smoker'].value_counts()
```

```python
insurance['smoker'].describe()
```

```python
charges = sns.boxplot(x = 'charges', y = 'smoker', data = insurance)
```
