```python
# Import package
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gapminder = pd.read_csv('Data Science/Country Project/Data/gapminder.csv')
gapminder
```

```python
gapminder = gapminder.drop(columns=['Unnamed: 0'])
gapminder
```

```python
# Scatter plot
plt.scatter(gapminder['gdp_cap'], gapminder['life_exp'])

# Put the x-axis on a logarithmic scale
plt.xscale('log')

# Show plot
plt.show()
```

```python
# Build Scatter plot
plt.scatter(gapminder['gdp_cap'],gapminder['life_exp'])

# Show plot
plt.show()
```

```python
# Create histogram
plt.hist(gapminder['life_exp'])

# Display histogram
plt.show()
```

```python
# Build histogram with 5 bins
plt.hist(gapminder['life_exp'],bins=5)

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 20 bins
plt.hist(gapminder['life_exp'],bins=20)

# Show and clean up again
plt.show()
plt.clf()
```

```python
# Histogram of life_exp, 15 bins
plt.hist(gapminder['life_exp'],bins=15)

# Show and clear plot
plt.show()
plt.clf()
```

```python
# Basic scatter plot, log scale
plt.scatter(gapminder['gdp_cap'], gapminder['life_exp'])
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)


# Add title
plt.title(title)

# After customizing, display the plot
plt.show()
```

```python
# Scatter plot
plt.scatter(gapminder['gdp_cap'], gapminder['life_exp'])

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val,tick_lab)


# After customizing, display the plot
plt.show()
```
